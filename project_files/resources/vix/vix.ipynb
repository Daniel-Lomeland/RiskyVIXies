```python
# Dependencies
import requests
import json
import pandas as pd
# from ratelimit import limits, sleep_and_retry
import time

# Retrieve Alpha Vantage key from config.py
from config import alpha_key
```

### Get 5 minute VIX data for the current day and store in a new df

```python
# API call to Alphavantage to get VIX data

url = f"https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=VIX&interval=5min&apikey={alpha_key}"
response = requests.get(url).json()
print(response)
```

```python
data_items = response['Time Series (5min)']
vix_df = pd.DataFrame.from_dict(data_items, orient='index')
vix_df.head()
```

```python
# Drop volume column and rename the column headers
clean_vix_df = vix_df.drop(['5. volume'], axis=1)
clean_vix_df.head()
```

```python
clean_vix_df.columns = ['open', 'high','low','close']
clean_vix_df.head()
```

```python
### Export to csv
clean_vix_df.to_csv("vix_intraday.csv", index=True, header=True)
```
