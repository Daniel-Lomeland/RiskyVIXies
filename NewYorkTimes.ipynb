```python
import json
import pandas as pd
import requests
import pprint
from config import api_key
import time
```

```python
#url_target = "https://newsapi.org/v2/top-headlines?"
year = '2019'
month = '3'
url_target = "https://api.nytimes.com/svc/archive/v1/"
query_url = f"{url_target}{year}/{month}.json?api-key={api_key}"
print(query_url)
```

```python
nyt_news=requests.get(query_url).json()
```

```python
nyt_headlines = []
nyt_date = []
nyt_len = len(nyt_news['response']['docs'])
```

```python
for i in range(nyt_len):
    print(f"Count: {i}")
    ns = nyt_news['response']['docs'][i]['snippet']
    print(ns)
    nyt_headlines.append(ns)
    
    nd = nyt_news['response']['docs'][i]['pub_date']
    print(nd)
    nyt_date.append(nd)
    
    print("---------------------------------------------------------------------------------------------------")
    
```

```python
nyt_news_df = pd.DataFrame({
    "Published" : nyt_date,
    "Headlines" : nyt_headlines
})

nyt_news_df.to_csv('NYT.csv', index=False)
```

```python
snippet = []
pubDate = []
```

```python
for i in range(101,105):
    url_target = 'https://api.nytimes.com/svc/search/v2/'
    query_url = f"{url_target}articlesearch.json?q=Business&page={i}&begin_date=20190101&end_date=20190301&api-key={api_key}"
    print(query_url)
    
    time.sleep(1)
    nyt_business_news=requests.get(query_url).json()
    
    news_len = len(nyt_business_news['response']['docs'])
    
    for nyt_len in range(news_len):
        pub_date = nyt_business_news['response']['docs'][nyt_len]['pub_date']
        snip = nyt_business_news['response']['docs'][nyt_len]['snippet']
        
        snippet.append(snip)
        pubDate.append(pub_date)
        
        print(f'Published Date: {pub_date}')
        print(f"Snippet: {snip}")
        print("\n")
        
    print("===================================================================================================")
        
    
```

```python
business_news_df = pd.DataFrame({
    "Published" : pubDate,
    "Headline"  : snippet
})

len(business_news_df)
```
